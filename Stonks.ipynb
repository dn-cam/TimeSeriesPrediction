```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

#for time series
from pandas import Series
```

```python
data_raw = pd.read_csv('data.csv')
```

### Sort the data by Date, and filter out only the data for the stock "GF"

```python
data_1 = data_raw
data_1['DATE'] = pd.to_datetime(data_1.DATE)
```

```python
data_1.sort_values(by='DATE')
```

```python
data_gf = data_1[data_1.TICKER == 'GF']
```

### Plotting the Autocorrelation Plot

```python
series = data_gf.set_index('DATE')
ts = series.cumsum()

ts.plot(figsize=(15, 10))
```

```python
plt.figure(figsize=(17, 8))
plt.plot(data_gf.DATE, data_gf.CLOSE)
plt.xlabel('Dates', fontsize=15)
plt.ylabel('Closing Price', fontsize=15)
plt.title("GF Stock Closing Prices")
```

```python
data_gf.head()
```

```python
drop_columns = ['TICKER', 'FIGI', 'TYPE', 'FREQUENCY', 'VOLUME', 'ADJ_OPEN', 'ADJ_HIGH', 'ADJ_LOW', 'ADJ_CLOSE', 'ADJ_VOLUME', 'ADJ_FACTOR', 'EX_DIVIDEND', 'SPLIT_RATIO']
data_gf.drop(drop_columns, axis = 1, inplace=True)
```

```python
data_gf.head()
series = data_gf.set_index('DATE')
ts = series.cumsum()

series.CLOSE.plot(figsize=(15, 10))
```

### Moving Average

```python
rolling_3 = series.rolling(window=3)
rolling_3_mean = rolling_3.mean()
print(rolling_3_mean.head(10))
# plot original and transformed dataset
series.CLOSE.plot(fig=(15, 10))
rolling_3_mean.CLOSE.plot(fig=(15, 10), color='red')
plt.show()
```

```python
rolling_5 = series.rolling(window=5)
rolling_5_mean = rolling_5.mean()

rolling_10 = series.rolling(window=10)
rolling_10_mean = rolling_10.mean()

rolling_15 = series.rolling(window=15)
rolling_15_mean = rolling_15.mean()

rolling_20 = series.rolling(window=20)
rolling_20_mean = rolling_20.mean()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,10))
plt.subplot(3, 2, 1)
series.CLOSE.plot()
plt.xlabel('window = 1')
plt.subplot(3, 2, 2)
rolling_3_mean.CLOSE.plot()
plt.xlabel('window = 3')
plt.subplot(3, 2, 3)
rolling_5_mean.CLOSE.plot()
plt.xlabel('window = 5')
plt.subplot(3, 2, 4)
rolling_10_mean.CLOSE.plot()
plt.xlabel('window = 10')
plt.subplot(3, 2, 5)
rolling_15_mean.CLOSE.plot()
plt.xlabel('window = 15')
plt.subplot(3, 2,6)
rolling_20_mean.CLOSE.plot()
plt.xlabel('window = 20')
fig.tight_layout()
```

Since the smoothing is not significant with windows of size even 20, we can try smoothing over a month, a quarter and a year.

```python
rolling_30 = series.rolling(window=30)
rolling_30_mean = rolling_30.mean()

rolling_90 = series.rolling(window=90)
rolling_90_mean = rolling_90.mean()

rolling_365 = series.rolling(window=365)
rolling_365_mean = rolling_365.mean()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
plt.subplot(2, 2, 1)
series.CLOSE.plot()
plt.xlabel('window = 1')
plt.subplot(2, 2, 2)
rolling_30_mean.CLOSE.plot()
plt.xlabel('window = 1 mo')
plt.subplot(2, 2, 3)
rolling_90_mean.CLOSE.plot()
plt.xlabel('window = 1 qt')
plt.subplot(2, 2, 4)
rolling_365_mean.CLOSE.plot()
plt.xlabel('window = 1 yr')
fig.tight_layout()
```
